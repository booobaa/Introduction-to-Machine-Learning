Scikit-learn s‚Äôappuie sur ces fonctionnalit√©s pour mettre en ≈ìuvre ses algorithmes. Parmi elles, le module `scipy.sparse` est particuli√®rement important : il permet de manipuler des matrices creuses, une repr√©sentation efficace des donn√©es largement utilis√©e dans scikit-learn.

Les matrices creuses sont utiles lorsqu‚Äôon doit stocker un tableau 2D compos√© en grande partie de valeurs nulles, car elles √©conomisent m√©moire et temps de calcul.

```python
from scipy import sparse

# Cr√©er un tableau NumPy 2D avec une diagonale de 1 et des 0 partout ailleurs
eye = np.eye(5)
print("NumPy array:\n", eye)
```

```python
# Convertir le tableau NumPy en une matrice creuse SciPy au format CSR
# Seules les valeurs non nulles sont conserv√©es
sparse_matrix = sparse.csr_matrix(eye)
print("\nSciPy sparse CSR matrix:\n", sparse_matrix)
```

Il n‚Äôest en pratique pas possible de transformer des donn√©es √©parses en repr√©sentations denses, car cela consommerait trop de m√©moire. La solution consiste donc √† travailler directement avec des repr√©sentations √©parses. Une mani√®re de le faire est d‚Äôutiliser le format COO (Coordinate list), qui permet de construire une matrice creuse en indiquant uniquement les positions et les valeurs des √©l√©ments non nuls.

### Explication

**Donn√©es √©parses** : ce sont des tableaux ou matrices qui contiennent surtout des z√©ros.

**Repr√©sentation dense** : stocker tous les √©l√©ments, y compris les z√©ros ‚Üí inefficace en m√©moire et en calcul.

**Repr√©sentation √©parse** : ne stocker que les valeurs non nulles et leurs indices ‚Üí beaucoup plus √©conomique.

**Format COO** : une des fa√ßons de repr√©senter une matrice creuse.

On garde trois vecteurs :

1. les valeurs non nulles,

2. les indices de ligne,

3. les indices de colonne.

Exemple : une matrice 3√ó3 avec seulement deux valeurs non nulles sera repr√©sent√©e par deux coordonn√©es et deux valeurs, au lieu de neuf cases.

```python
import numpy as np
from scipy.sparse import coo_matrix

# Cr√©ation d'une matrice dense
dense = np.array([
    [0, 0, 3],
    [4, 0, 0],
    [0, 0, 5]
])

# Conversion en matrice creuse au format COO
sparse_coo = coo_matrix(dense)

print("Matrice COO :")
print(sparse_coo)

# Affichage des donn√©es internes
print("\nValeurs non nulles :", sparse_coo.data)
print("Indices de lignes :", sparse_coo.row)
print("Indices de colonnes :", sparse_coo.col)
```

#### Autre exemple:

```python
data = np.ones(4)
row_indices = np.arange(4)
col_indices = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print("COO representation:\n", eye_coo)
```

# √âvaluation d‚Äôun mod√®le de classification des iris: donn√©es d'entra√Ænement et de test

## Objectif
Nous voulons √©laborer un mod√®le d‚Äôapprentissage automatique √† partir de ces donn√©es, afin de pr√©dire l‚Äôesp√®ce d‚Äôiris correspondant √† de nouvelles mesures. Toutefois, avant de l‚Äôappliquer √† des observations in√©dites, il est essentiel de v√©rifier sa validit√©, c‚Äôest-√†-dire la fiabilit√© de ses pr√©dictions.

## Limite de l‚Äô√©valuation sur l‚Äôensemble d‚Äôentra√Ænement
Il n‚Äôest pas possible d‚Äô√©valuer le mod√®le avec les donn√©es qui ont servi √† son entra√Ænement.  
- Le mod√®le peut simplement ¬´‚ÄØm√©moriser‚ÄØ¬ª l‚Äôensemble d‚Äôapprentissage.  
- Il fournirait alors syst√©matiquement la bonne √©tiquette pour ces points.  
- Cela ne garantit pas sa capacit√© √† **g√©n√©raliser** √† des donn√©es nouvelles.

## Strat√©gie d‚Äô√©valuation
Pour mesurer ses performances, on lui pr√©sente des donn√©es qu‚Äôil n‚Äôa jamais vues, mais dont les √©tiquettes sont connues.  
La pratique courante consiste √† diviser l‚Äôensemble initial de donn√©es √©tiquet√©es (ici, 150 mesures de fleurs) en deux parties‚ÄØ:

- **Ensemble d‚Äôentra√Ænement** : utilis√© pour construire le mod√®le.  
- **Ensemble de test (ou de validation)** : utilis√© pour √©valuer la performance du mod√®le.  

## Outil scikit-learn
La biblioth√®que **scikit-learn** propose la fonction `train_test_split` :  
- Elle m√©lange les donn√©es.  
- Elle extrait **75‚ÄØ%** des lignes pour l‚Äôentra√Ænement, avec leurs √©tiquettes correspondantes.  
- Les **25‚ÄØ%** restants constituent l‚Äôensemble de test.  

Bien que le choix de cette proportion soit arbitraire, utiliser environ un quart des donn√©es pour le test est une bonne pratique.

## Conventions de notation
Dans scikit-learn :  
- Les donn√©es sont not√©es **X** (majuscule).  
- Les √©tiquettes sont not√©es **y** (minuscule).  

Cette convention s‚Äôinspire de la formulation math√©matique :  



\[
f(x) = y
\]



o√π \(x\) repr√©sente l‚Äôentr√©e et \(y\) la sortie.  
- \(X\) majuscule ‚Üí matrice de donn√©es (tableau bidimensionnel).  
- \(y\) minuscule ‚Üí vecteur cible (tableau unidimensionnel).

Appelons la fonction train_test_split sur nos donn√©es et assignons les sorties selon cette nomenclature¬†:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

# Fonction `train_test_split` et m√©lange des donn√©es

## M√©lange pr√©alable
Avant de s√©parer les donn√©es, la fonction `train_test_split` les m√©lange √† l‚Äôaide d‚Äôun g√©n√©rateur de nombres pseudo-al√©atoires.  
- Sans ce m√©lange, si l‚Äôon prenait simplement les 25‚ÄØ% de lignes restantes comme ensemble de test, toutes appartiendraient √† la classe **2**, car les donn√©es sont ordonn√©es par √©tiquette (voir `iris['target']`).  
- Un ensemble de test limit√© √† une seule classe ne permettrait pas d‚Äô√©valuer correctement la capacit√© de **g√©n√©ralisation** du mod√®le.  
- Le m√©lange garantit que l‚Äôensemble de test contient des exemples issus de toutes les classes.

## Param√®tre `random_state`
Pour obtenir un r√©sultat identique lors de plusieurs ex√©cutions :  
- On fixe une graine au g√©n√©rateur pseudo-al√©atoire via le param√®tre `random_state`.  
- Le processus devient **d√©terministe**‚ÄØ: la m√™me ligne de code produira toujours le m√™me d√©coupage.  
- Dans ce cours, nous utiliserons syst√©matiquement `random_state` pour toutes les proc√©dures impliquant de l‚Äôal√©atoire.

## R√©sultats de la fonction
La fonction `train_test_split` retourne quatre tableaux NumPy :  
- `X_train` ‚Üí 75‚ÄØ% des lignes du jeu de donn√©es.  
- `X_test` ‚Üí les 25‚ÄØ% restantes.  
- `y_train` et `y_test` ‚Üí les √©tiquettes correspondantes.

#### Cr√©ation de votre premier mod√®le : k plus proches voisins (k-Nearest Neighbors)

Nous allons maintenant construire un premier mod√®le d‚Äôapprentissage automatique.  
La biblioth√®que **scikit-learn** propose de nombreux algorithmes de classification, et nous allons utiliser ici celui des **k plus proches voisins (k-NN)**, r√©put√© pour sa simplicit√©.

##### Principe de fonctionnement

- Le mod√®le ne fait qu‚Äô**enregistrer l‚Äôensemble d‚Äôentra√Ænement**.  
- Lorsqu‚Äôun nouveau point de donn√©es doit √™tre class√© :
  1. L‚Äôalgorithme recherche le ou les points les plus proches dans l‚Äôensemble d‚Äôentra√Ænement.  
  2. Il attribue au nouveau point l‚Äô√©tiquette de ces voisins.

##### Le r√¥le de *k*

- Par d√©faut, on pourrait n‚Äôutiliser que **le voisin le plus proche**.  
- Mais il est souvent pr√©f√©rable de consid√©rer un nombre fixe de voisins (*k*), par exemple 3 ou 5.  
- La pr√©diction est alors bas√©e sur la **classe majoritaire** parmi ces voisins.  
- Dans notre premier exemple, nous nous limiterons √† **un seul voisin** (*k = 1*).  
- Ce point sera approfondi au chapitre 2.

##### Impl√©mentation avec scikit-learn

- Tous les mod√®les de scikit-learn sont regroup√©s dans des **classes appel√©es Estimator**.  
- L‚Äôalgorithme k-NN est impl√©ment√© dans la classe **`KNeighborsClassifier`** du module `neighbors`.  
- Pour l‚Äôutiliser :
  1. On instancie la classe.  
  2. On d√©finit ses param√®tres, dont le plus important est **le nombre de voisins**.  
  3. Dans notre cas, nous fixons ce param√®tre √† **1**.


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

# L‚Äôobjet knn et la construction du mod√®le

L‚Äôobjet **knn** regroupe l‚Äôalgorithme utilis√© pour :
- b√¢tir le mod√®le √† partir des donn√©es d‚Äôentra√Ænement,  
- r√©aliser des pr√©dictions sur de nouvelles observations.  

Il conserve √©galement les informations issues de l‚Äôensemble d‚Äôentra√Ænement.  
Dans le cas de **`KNeighborsClassifier`**, cela signifie qu‚Äôil se contente de stocker les donn√©es d‚Äôentra√Ænement.

## Mise en place du mod√®le

Pour entra√Æner le mod√®le, on utilise la m√©thode **`fit`** de l‚Äôobjet knn :  
- Elle prend en entr√©e deux tableaux NumPy :  
  - **`X_train`** : les donn√©es d‚Äôentra√Ænement,  
  - **`y_train`** : les √©tiquettes correspondantes.  
- L‚Äôappel √† `fit` permet d‚Äôassocier les donn√©es aux labels afin que le mod√®le soit pr√™t √† effectuer des pr√©dictions.


```python
knn.fit(X_train, y_train)
```

La m√©thode `fit` renvoie l'objet knn lui-m√™me (et le modifie sur place), ce qui nous donne une repr√©sentation sous forme de cha√Æne de caract√®res de notre classificateur. Cette repr√©sentation indique les param√®tres utilis√©s pour cr√©er le mod√®le. Presque tous sont les valeurs par d√©faut, mais on peut trouver `n_neighbors=1`, qui est le param√®tre que nous avons pass√©. La plupart des mod√®les de scikit-learn ont de nombreux param√®tres, mais la majorit√© d'entre eux servent √† optimiser la vitesse ou sont destin√©s √† des cas d'utilisation tr√®s sp√©cifiques. Vous n‚Äôavez pas √† vous soucier des autres param√®tres affich√©s ici. L‚Äôaffichage d‚Äôun mod√®le scikit-learn peut g√©n√©rer des cha√Ænes de caract√®res tr√®s longues, mais ne vous laissez pas intimider.

```python
#### Faire des Predictions

Nous pouvons d√©sormais effectuer des pr√©dictions √† l'aide de ce mod√®le sur de nouvelles donn√©es dont nous ignorons peut-√™tre les √©tiquettes exactes. Imaginons que nous trouvions un iris sauvage pr√©sentant des s√©pales de 5 cm de long et 2,9 cm de large, ainsi que des p√©tales de 1 cm de long et 0,2 cm de large.

√Ä quelle esp√®ce d'iris cela correspond-il¬†?

Nous pouvons int√©grer ces donn√©es dans un tableau NumPy, en calculant sa forme, c'est-√†-dire le nombre d'√©chantillons (1) multipli√© par le nombre de caract√©ristiques (4).
```

#### √âvaluation du mod√®le

C‚Äôest ici qu‚Äôintervient l‚Äôensemble de test cr√©√© pr√©c√©demment. Ces donn√©es n‚Äôont pas servi √† construire le mod√®le, mais nous connaissons l‚Äôesp√®ce de chaque iris de l‚Äôensemble de test.

Pour ce mod√®le, la pr√©cision sur l'ensemble de test est d'environ 0,97, ce qui signifie que nous avons correctement pr√©dit la couleur des iris pour 97 % d'entre eux.
